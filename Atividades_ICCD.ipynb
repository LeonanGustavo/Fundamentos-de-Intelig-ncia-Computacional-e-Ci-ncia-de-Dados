**A:** sample distribution and sampling distribution

```python
#import das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1: Criar a população artificial (onde irei trbalhar com notas ENEM)
np.random.seed(42) # seed garante que quando o cod for executado novamente seja gerado os mesmo numeros
populacao = np.random.normal(loc=600, scale=100, size=10000)  # média 600, desvio 100
```

```python
# 2: Visualizar a distribuição da população
plt.figure(figsize=(10, 4))
sns.histplot(populacao, bins=50, kde=True, color='blue') # Bins = Número de colunas no histograma, "kde=True" desenha a curva suavizada da distribuição.
plt.title("Distribuição da População de Notas (ENEM)")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# 3: Simular 200 amostras de 30 alunos
n_amostras = 200
tamanho_amostra = 30
medias_amostrais = []
```

```python
for _ in range(n_amostras):
    amostra = np.random.choice(populacao, size=tamanho_amostra) # sorteia 30 alunos da população.
    medias_amostrais.append(np.mean(amostra)) # calcula a média das notas da amostra e salva essa média na lista.
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(medias_amostrais, bins=30, kde=True, color='orange')
plt.title("Distribuição das Médias Amostrais (sampling distribution)")
plt.xlabel("Média das Notas")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# Avaliações de Probabilidade de Eventos

from scipy.stats import norm

# Parâmetros da população
media_pop = np.mean(populacao)
desvio_pop = np.std(populacao)
```

```python
# 1: Probabilidade de um aluno ter nota maior que 700
prob_maior_700 = 1 - norm.cdf(700, loc=media_pop, scale=desvio_pop)
print(f"Probabilidade de um aluno ter nota > 700: {prob_maior_700:.2%}")
```

```python
# Criar a população assimétrica
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gamma
np.random.seed(42) #
```

```python
# Simular população de notas ENEM com distribuição Gama
populacao_enem_gama = gamma.rvs(a=5, scale=80, size=10000)  # média ≈ a * scale = 400
```

```python
# Visualizar a distribuição da população
plt.figure(figsize=(10, 4))
sns.histplot(populacao_enem_gama, bins=50, kde=True, color='blue')
plt.title("Distribuição de Notas ENEM - População (Distribuição Gama)")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# criação função de simulação de medidas amostrais
def simular_medias_amostrais(populacao, n, n_amostras=1000):
    medias = []
    for _ in range(n_amostras):
        amostra = np.random.choice(populacao, size=n)
        medias.append(np.mean(amostra))
    return medias
```

```python
medias = simular_medias_amostrais(populacao_enem_gama, n=10)
print(np.round(medias, 2))
```
