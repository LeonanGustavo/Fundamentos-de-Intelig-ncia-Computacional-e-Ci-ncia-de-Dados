**A:** sample distribution and sampling distribution

```python
#import das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1: Criar a população artificial (onde irei trbalhar com notas ENEM)
np.random.seed(42) # seed garante que quando o cod for executado novamente seja gerado os mesmo numeros
populacao = np.random.normal(loc=600, scale=100, size=10000)  # média 600, desvio 100
```

```python
# 2: Visualizar a distribuição da população
plt.figure(figsize=(10, 4))
sns.histplot(populacao, bins=50, kde=True, color='blue') # Bins = Número de colunas no histograma, "kde=True" desenha a curva suavizada da distribuição.
plt.title("Distribuição da População de Notas (ENEM)")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# 3: Simular 200 amostras de 30 alunos
n_amostras = 200
tamanho_amostra = 30
medias_amostrais = []
```

```python
for _ in range(n_amostras):
    amostra = np.random.choice(populacao, size=tamanho_amostra) # sorteia 30 alunos da população.
    medias_amostrais.append(np.mean(amostra)) # calcula a média das notas da amostra e salva essa média na lista.
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(medias_amostrais, bins=30, kde=True, color='orange')
plt.title("Distribuição das Médias Amostrais (sampling distribution)")
plt.xlabel("Média das Notas")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# Avaliações de Probabilidade de Eventos

from scipy.stats import norm

# Parâmetros da população
media_pop = np.mean(populacao)
desvio_pop = np.std(populacao)
```

```python
# 1: Probabilidade de um aluno ter nota maior que 700
prob_maior_700 = 1 - norm.cdf(700, loc=media_pop, scale=desvio_pop)
print(f"Probabilidade de um aluno ter nota > 700: {prob_maior_700:.2%}")
```

```python
# criação função de simulação de medidas amostrais
def simular_medias_amostrais(populacao, n, n_amostras=1000):
    medias = []
    for _ in range(n_amostras):
        amostra = np.random.choice(populacao, size=n)
        medias.append(np.mean(amostra))
    return medias
```

**C:** Cálculo de p-valor com visualização

image.png

```python
# Dados do experimento
n = 6  # número total de observações
k = 1  # número de sucessos observados (vermelhos)
p_hipotese = 0.5  # hipótese nula: chance de vermelho é 50%
```

```python
from scipy.stats import binom

# Calcular p-valor: P(X <= 1) + P(X >= 5) [teste bicaudal] Verificar se o valor observado é “extremo” de ambos os lados.
p_valor = binom.cdf(k, n, p_hipotese) + (1 - binom.cdf(n - k - 1, n, p_hipotese))
print(f"P-valor: {p_valor:.4f}")
```

Se o p-valor < 0.05, rejeitamos a hipótese nula (não parece ser puro acaso).

Se o p-valor >= 0.05, não rejeitamos — como pode acontecer aqui.

```python
import matplotlib.pyplot as plt
import numpy as np

# Distribuição binomial
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p_hipotese)

plt.figure(figsize=(8, 5))
bars = plt.bar(x, y, color='blue')
bars[k].set_color('red')  # destaque no valor observado
bars[n - k].set_color('red')  # destaque no outro lado extremo
plt.title("Distribuição Binomial: P(x) com n= 6 e p= 0.5")
plt.xlabel("Número de eventos vermelhos")
plt.ylabel("Probabilidade")
plt.grid(axis='y')
plt.show()
```

**D:** Cálculo de p-value entre duas distribuições Gaussianas


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definição dos parâmetros
mu_vermelho = 0       # Média da hipótese nula (distribuição vermelha)
mu_azul = 1      # Média da hipótese alternativa (distribuição azul)
sigma = 1     # Desvio padrão (igual nas duas)

# Intervalo de x para visualização das curvas
x = np.linspace(-4, 5, 500)

# Densidade de probabilidade das duas normais
y_vermelho = norm.pdf(x, loc=mu_vermelho, scale=sigma)
y_azul = norm.pdf(x, loc=mu_azul, scale=sigma)

# Plotando as duas distribuições
plt.figure(figsize=(10, 4))
plt.plot(x, y_vermelho, 'r-', label='H₀ (vermelha)')  # curva da hipótese nula
plt.plot(x, y_azul, 'b-', label='H₁ (azul)')          # curva da hipótese alternativa
plt.axvline(x=mu_vermelho, color='red', linestyle='--')  # marca a média da H₀
plt.axvline(x=mu_azul, color='blue', linestyle='--')     # marca a média da H₁
plt.legend()
plt.title("Distribuições Normal: H0 (vermelha) vs H1 (azul)")
plt.xlabel("Valor")
plt.ylabel("Densidade de Probabilidade")
plt.grid(True)
plt.show()
```

```python
# Definir semente para reprodutibilidade
np.random.seed(32)

# Parâmetros da simulação
n_amostras = 100           # número de sequências
tamanho_amostra = 3        # observações
p_valores = []             # lista para armazenar os p-valores calculados

for _ in range(n_amostras):

    amostra = np.random.normal(loc=mu_azul, scale=sigma, size=tamanho_amostra) # Gera uma amostra com 3 valores da distribuição azul (H1)

    media_amostra = np.mean(amostra)

    z = (media_amostra - mu_vermelho) / (sigma / np.sqrt(tamanho_amostra)) # Calcular o valor z comparando com H0

    p = 2 * (1 - norm.cdf(abs(z)))  # Calcular o p-valor bilateral

    p_valores.append(p)
```

```python
# Conta quantos p-valores são maiores que 0.05, Isso indica que não rejeitamos H0, mesmo ela sendo falsa
falsos_negativos = sum(p > 0.05 for p in p_valores)

taxa_falso_negativo = falsos_negativos / n_amostras # Calcula a taxa: porcentagem de falsos negativos

print(f"Taxa de falsos negativos (p > 0.05): {taxa_falso_negativo:.2%}")
```
