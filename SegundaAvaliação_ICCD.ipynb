```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df_clean = df[df['totmembers'].notnull() & (df['totmembers'] > 0)] # Remove valores nulos (notnull()) e Garante que só há valores positivos de membros (> 0)
```

```python
X = df_clean[['totmembers']]
y = df_clean['mdeaths']

```

```python
model = LinearRegression()
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_
#Cria e treina o modelo de regressão linear.
#Obtém os dois parâmetros principais:
#intercept: valor de y quando x = 0.
#slope: coeficiente angular da reta (inclinação).
```

```python
X_sm = sm.add_constant(X) #statsmodels exige que o vetor X tenha uma constante explícita
model_sm = sm.OLS(y, X_sm).fit()
regression_summary = model_sm.summary()
#Calcula o modelo OLS (mínimos quadrados).
#summary() gera um relatório estatístico completo da regressão.
```

```python
x_line = np.linspace(X.min(), X.max(), 100)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='brown', edgecolor='k', label="Dados")

y_line = intercept + slope * x_line
plt.plot(x_line, y_line, color='black', linewidth=3, label=f"Regressão real: y = {intercept:.2f} + {slope:.2f}x")
plt.xlabel('Total de Membros')
plt.ylabel('Mortes (mdeaths)')
```

```python
intercepts = [-1, 0, 0.5, 1, 2, 4] # Valores de a testados
```

```python
# Coeficiente angular (slope) da regressão linear ajustada
slope = model.coef_[0]

# Eixo X para as linhas
x_vals = np.linspace(df_clean['totmembers'].min(), df_clean['totmembers'].max(), 100)
```

```python
a = -1
y_vals = a + slope * x_vals

plt.figure(figsize=(6, 4))
plt.scatter(df_clean['totmembers'], df_clean['mdeaths'], color='darkred', alpha=0.6, label='Dados Reais')
plt.plot(x_vals, y_vals, color='black', linewidth=2, label=f'y = {a} + {slope:.2f}x')
plt.title('Case 1: a = -1')
plt.xlabel('Total de Membros')
plt.ylabel('Mortes (mdeaths)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
a = 0
y_vals = a + slope * x_vals

plt.figure(figsize=(6, 4))
plt.scatter(df_clean['totmembers'], df_clean['mdeaths'], color='darkred', alpha=0.6, label='Dados Reais')
plt.plot(x_vals, y_vals, color='black', linewidth=2, label=f'y = {a} + {slope:.2f}x')
plt.title('Case 2: a = 0')
plt.xlabel('Total de Membros')
plt.ylabel('Mortes (mdeaths)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score


# Valores de a para testar
intercepts = [-1, 0, 0.5, 1, 2, 4]
slope = model.coef_[0]  # slope fixo aprendido pelo modelo
X = df_clean['totmembers'].values
y_true = df_clean['mdeaths'].values

# R² do modelo original
r2_sklearn = model.score(df_clean[['totmembers']], df_clean['mdeaths'])

print(f'R² do modelo original (scikit-learn): {r2_sklearn:.4f}\n')

# Cálculo manual de R² para cada a
for a in intercepts:
    y_pred = a + slope * X
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    print(f'a = {a:<4} | R² = {r2:.4f}')
```

   **Análise resultado**
 Após aplicar a regressão linear com o scikit-learn, obtive um R² de 0.0212 — ou seja, o modelo explica cerca de 2% da variação nos dados. A seguir, testei seis valores diferentes para o intercepto a, mantendo o coeficiente original, para observar como isso afeta o ajuste.

Os resultados foram claros: qualquer desvio do intercepto ideal prejudica bastante o modelo. Por exemplo, quando usei a = 0, o R² caiu ligeiramente para 0.0199. Já com valores como a = 1 ou a = 2, o R² ficou negativo, o que significa que esses modelos são piores do que simplesmente prever a média dos dados.

Isso mostra na prática que o ajuste feito automaticamente pelo scikit-learn encontra os melhores parâmetros, e que mudar o intercepto sem critério pode comprometer seriamente a qualidade da regressão
